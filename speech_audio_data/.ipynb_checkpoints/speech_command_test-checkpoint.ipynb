```python
import warnings; warnings.simplefilter('ignore')

from scipy.io import wavfile
import pandas as pd
import numpy as np
from plotnine import *
theme_set(theme_bw())
```

```python
sample_rate, dog = wavfile.read('./sounds/dog/00b01445_nohash_0.wav')
dog_pd = pd.DataFrame(dog, columns=['amplitude']).assign(time_id = range(len(dog)))
dog_pd.head()
```

```python
(
    ggplot(dog_pd) 
      + geom_line(aes(x = 'time_id', 
                      y = 'amplitude'))
      + theme(figure_size=(4,2))
)
```
